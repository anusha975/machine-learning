```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y):
    m_curr = b_curr = 0
    rate = 0.01
    n = len(x)
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(10000):
        y_predicted = m_curr * x + b_curr
#         print (m_curr,b_curr, i)
        plt.plot(x,y_predicted,color='green')
        md = -(2/n)*sum(x*(y-y_predicted))
        yd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gradient_descent_animated(x, y, iterations=50, learning_rate=0.01):
    m_curr = b_curr = 0
    n = len(x)

    plt.scatter(x, y, color='red', marker='+', s=100)  # original data

    for i in range(iterations):
        y_pred = m_curr * x + b_curr
        plt.plot(x, y_pred, color='green', alpha=0.3)  # each step
        # Gradient descent update
        md = -(2/n) * sum(x * (y - y_pred))
        bd = -(2/n) * sum(y - y_pred)
        m_curr -= learning_rate * md
        b_curr -= learning_rate * bd

    # Final line in bold
    plt.plot(x, m_curr * x + b_curr, color='blue', linewidth=2)
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.title("Gradient Descent Steps (green) and Final Fit (blue)")
    plt.show()

    print(f"Final m: {m_curr:.2f}, b: {b_curr:.2f}")

# Data
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent_animated(x, y, iterations=50, learning_rate=0.01)
```
